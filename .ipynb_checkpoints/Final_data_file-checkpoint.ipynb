```python
# 4. apartment_raw: 
# the raw data of apartments information in Pittsburgh
titles=list()
addresses=list()
zipcodes=list()
prices=list()
for num in range(1,29):
    #data retrive from this website
    page=requests.get('https://www.apartments.com/pittsburgh-pa/%s/?bb=z1nsvpj4rIpt24vwa'%num)
    page=page.content
    soup = bs(page, 'html.parser')
    anchors=soup.find_all('a',class_='placardTitle js-placardTitle ')
    anchors2=soup.find_all('div',class_='location')
    for a in anchors:
        if a.has_attr('title'):
            title=a.get('title')
            titles.append(title)
    for a in anchors2:        
        if a.has_attr('title'):
            address=a.get('title')
            addresses.append(address)
            zipcodes.append(address)
apt_raw=list()
apt_raw.append(titles)
apt_raw.append(addresses)
apt_raw.append(prices)
apt_raw.append(zipcodes)
print(apt_raw)
```

```python
# 5. apartment_cleaned: 
# the cleaned data of apartments information in Pittsburgh
titles=list()
addresses=list()
zipcodes=list()
prices=list()
for num in range(1,29):
    #data retrive from this website
    page=requests.get('https://www.apartments.com/pittsburgh-pa/%s/?bb=z1nsvpj4rIpt24vwa'%num)
    page=page.content
    soup = bs(page, 'html.parser')
    anchors=soup.find_all('a',class_='placardTitle js-placardTitle ')
    anchors2=soup.find_all('div',class_='location')
    for a in anchors:
        if a.has_attr('title'):
            title=a.get('title')
            titles.append(title[:-16])
    for a in anchors2:        
        if a.has_attr('title'):
            address=a.get('title')
            addresses.append(address[:-22])
            zipcodes.append(address[-5:])
    anchor3=soup.find_all('span',class_='altRentDisplay')
    for i in anchor3:
        i=i.text.strip()
        if i.find('-')!=-1:
            prices.append(i[1:6])
        else:
            prices.append(i[1:])

apt_list = pd.DataFrame(
    {
     'name': titles,
     'zipcode': zipcodes,
     'address': addresses,
     'price':prices
    })
apt_list["price"] = apt_list["price"].str.replace("-", "")
apt_list['price'] = apt_list['price'].str.replace(',','')
apt_list['price'] = apt_list['price'].str.replace(' ','')
filter1 = apt_list['price'].str.contains('allforRent')
apt_list = apt_list[~filter1]

apt_list['price'] = pd.to_numeric(apt_list['price'])
apt_list["address"] = np.where(apt_list["address"]=="",apt_list["name"],apt_list["address"])
shady1=apt_list[apt_list.zipcode=='15213']
shady2=apt_list[apt_list.zipcode=='15232']
shady3=[shady1,shady2]
shady_list=pd.concat(shady3)
oak_list=apt_list[apt_list.zipcode=='15213']
town_list=apt_list[apt_list.zipcode=='15222']
hill1=apt_list[apt_list.zipcode=='15217']
hill2=apt_list[apt_list.zipcode=='15232']
hill3=[hill1,hill2]
hill_list=pd.concat(hill3)
display(apt_list)
```

```python
# 6. new_crime_raw: 
# raw new crime json data from spotcrime.com

with open('spotcrime.json', encoding='utf-8') as data_file:
    crime_parsed = json.loads(data_file.read())
print(crime_parsed)

print(crime_parsed)
data_file.close()

crimes = crime_parsed['crimes']
print(crimes)

crime_csv = open('spotcrime.csv', 'w')
csvwriter = csv.writer(crime_csv)

count = 0
for crime in crimes:
    if count == 0:
        header = crime.keys()
        csvwriter.writerow(header)
        count = count + 1
    csvwriter.writerow(crime.values())
crime_csv.close()
```

```python
# 9. historical_crime_cleaned: 
# cleaned historical crime data
old_data = pd.read_csv("historical_data.csv", low_memory=False)
old_data=old_data.drop(['PK', 'CCR','AGE','GENDER','RACE','ARRESTLOCATION','INCIDENTZONE','INCIDENTTRACT','COUNCIL_DISTRICT','PUBLIC_WORKS_DIVISION','INCIDENTNEIGHBORHOOD'], axis=1)
old_data = old_data.rename(columns={'_id': 'original_id', 'ARRESTTIME': 'date','OFFENSES':'type','INCIDENTLOCATION':'address','X':'lat','Y':'lon'})
old_data['zipcode']=old_data['address']
old_data['zipcode'] = old_data['zipcode'].astype(str).str[-5:]
old_data.loc[old_data['type'].str.contains('Assault'), 'type'] = 'Assult'
old_data.loc[old_data['type'].str.contains('Theft'), 'type'] = 'Theft'
old_data.loc[old_data['type'].str.contains('Assult|Theft')==False,'type']='Other'
old_data['address'] = old_data['address'].astype(str).str[:-21]
# data['zipcode']=pd.to_numeric(data['zipcode'], errors='coerce')
old_data.sort_values(by=['zipcode'])
old_data=old_data[old_data['zipcode'].apply(lambda x : x.isalnum())]
zipcode_list=old_data.zipcode.unique()
old_data = old_data[['original_id', 'type','date','address','lat','lon','zipcode']]
print(old_data)
```

```python
# 10. merged:
# merged cleaned data

merged_data= pd.concat([old_data,new_data])
pd.options.display.max_columns = None
display(merged_data)
```
